### 3. Using Pandas `pivot_table()` method, produce the same outcome as above (using manual groupby)

```python
pt = pd.pivot_table(neighbor_crime, values = 'Neighborhood',
                    columns = 'District',
                    index = 'Description',
                    aggfunc = 'count')
pt
```
